```python
label = ['슬픔','행복','놀람','분노','혐오','중립','공포']
```

```python
colors = ['lemonchiffon','pink','salmon','greenyellow','paleturquoise','lightgray','darkseagreen']
```

```python
p = round(df[df.song == '라라라'].sentiment.value_counts(normalize=True) * 100)
p.plot.pie(autopct='%.1f%%', colors=[colors[1],colors[0],colors[2],colors[5]])
plt.title('SG 워너비 - 라라라')
plt.legend()
plt.ylabel('')
plt.show()
```

```python
p = round(df[df.song == '좋은 사람 (Feat. 김형중)'].sentiment.value_counts(normalize=True) * 100)
p.plot.pie(autopct='%.1f%%', colors=[colors[1],colors[0],colors[6]])
plt.title('토이 - 좋은 사람 (Feat. 김형중)')
plt.legend()
plt.ylabel('')
plt.show()
```

```python
p = round(df[df.song == '편지 (Original Ver.)'].sentiment.value_counts(normalize=True) * 100)
p.plot.pie(autopct='%.1f%%', colors=[colors[0],colors[3],colors[5],colors[4],colors[1]])
plt.title('최재훈 - 편지 (Original Ver.)')
plt.legend()
plt.ylabel('')
plt.show()
```

```python
p = round(df[df.song == 'Monologue'].sentiment.value_counts(normalize=True) * 100)
p.plot.pie(autopct='%.1f%%', colors=[colors[0],colors[1],colors[6],colors[3]])
plt.title('테이 - Monologue')
plt.legend()
plt.ylabel('')
plt.show()
```

```python
p = round(df[df.song == '당신은 모르실거야 (CF - LIG 핑클편)'].sentiment.value_counts(normalize=True) * 100)
p.plot.pie(autopct='%.1f%%', colors=[colors[0],colors[6],colors[3],colors[1]])
plt.title('핑클 (Fin.K.L) - 당신은 모르실거야 (CF - LIG 핑클편)')
plt.legend()
plt.ylabel('')
plt.show()
```
